```python
from datasets import list_datasets
from tqdm import tqdm
from datasets import list_datasets, load_dataset
import pandas as pd
import json
```

```python
## load dataset
dataset = load_dataset('spider')
```

```python
dataset
```

```python
## iterate over train set and create question/sql pairs
train_set = []
for row in tqdm(dataset['train']):
    data = {
        'db_id': row['db_id'],
        'question': row['question'],
        'sql': row['query'],
        'query_toks': row['query_toks'],
        'query_toks_no_value': row['query_toks_no_value'],
        'question_toks': row['question_toks'],
    }
    train_set.append(data)
```

```python
## iterate over train set and create question/sql pairs
validation_set = []
for row in tqdm(dataset['train']):
    data = {
        'db_id': row['db_id'],
        'question': row['question'],
        'sql': row['query'],
        'query_toks': row['query_toks'],
        'query_toks_no_value': row['query_toks_no_value'],
        'question_toks': row['question_toks'],
    }
    validation_set.append(data)
```

```python
train_df, validation_df = pd.DataFrame(train_set), pd.DataFrame(validation_set)
```

```python
train_df.to_csv('data/spider/spider_schema_train.csv', index=False), validation_df.to_csv('data/spider/spider_schema_validation.csv', index=False)
```
